```python
#import 자료 및 url, htmlText

import requests
from bs4 import BeautifulSoup

url = "https://ko.wikipedia.org/wiki/%EC%98%AC%ED%95%B4%EC%9D%98_%EA%B2%8C%EC%9E%84_%EC%83%81_%EB%AA%A9%EB%A1%9D"

try:
    import Tkinter as tk
except:
    import tkinter as tk
    
from tkinter import Tk, ttk, Label, Button, Text, END


from tkinter import *
import tkinter as tk

# 메소드 모음

def call_Ign():
    title_feature.pack_forget()
    setTableItemIGN()
     
def callDest():
    title_feature.pack_forget()
    setTableItemDest()
    
def callEdge():
    title_feature.pack_forget()
    setTableItemEdge()

def callGiant():
    title_feature.pack_forget()
    setTableItemGiant()
  

    
def setTableItemIGN() :
    global title_feature
    window.title("IGN Goty 차트")
    title_feature = Label(window, text = "IGN Goty 차트", font = ("Bahnschrift", 20))
    title_feature.pack(padx = 10, pady = 15)
    title_feature['bg'] = 'sky blue'
    treeTable.delete(*treeTable.get_children())
    for idx, report in enumerate(gotyIgn) :
        year = report['year']
        game = report['game']
        genre = report['genre']
        flatform = report['flatform']
        developer = report['developer']
        treeTable.insert("", 'end', iid = None, text = str(idx), values = [year, game, genre, flatform, developer])
        
        
def setTableItemDest() :
    global title_feature
    window.title("디스트럭토이드 차트")
    title_feature = Label(window, text = "Destructoid Goty 차트", font = ("Bahnschrift", 20))
    title_feature.pack(padx = 10, pady = 15)
    title_feature['bg'] = 'sky blue'
    treeTable.delete(*treeTable.get_children())
    for idx, report in enumerate(gotyDest) :
        year = report['year']
        game = report['game']
        genre = report['genre']
        flatform = report['flatform']
        developer = report['developer']
        treeTable.insert("", 'end', iid = None, text = str(idx), values = [year, game, genre, flatform, developer])

        
def setTableItemEdge() :
    global title_feature
    window.title("에지 차트")
    title_feature = Label(window, text = "Edge Goty 차트", font = ("Bahnschrift", 20))
    title_feature.pack(padx = 10, pady = 15)
    title_feature['bg'] = 'sky blue'
    treeTable.delete(*treeTable.get_children())
    for idx, report in enumerate(gotyEdge) :
        year = report['year']
        game = report['game']
        genre = report['genre']
        flatform = report['flatform']
        developer = report['developer']
        treeTable.insert("", 'end', iid = None, text = str(idx), values = [year, game, genre, flatform, developer])

        
def setTableItemGiant() :
    global title_feature
    window.title("자이언트 밤 차트")
    title_feature = Label(window, text = "Giant Bomb Goty 차트", font = ("Bahnschrift", 20))
    title_feature.pack(padx = 10, pady = 15)
    title_feature['bg'] = 'sky blue'
    treeTable.delete(*treeTable.get_children())
    for idx, report in enumerate(goty_Giant_Bomb) :
        year = report['year']
        game = report['game']
        genre = report['genre']
        flatform = report['flatform']
        developer = report['developer']
        treeTable.insert("", 'end', iid = None, text = str(idx), values = [year, game, genre, flatform, developer])


# IGN 차트

htmlText = requests.get(url).text

bsoup = BeautifulSoup(htmlText, "html.parser")
btab = bsoup.find("table", {"class" : "wikitable"})

btrs = btab.find("tbody").find_all("tr")
btdcols = btrs[1].find_all("td")

gotyIgn = []

for tr in btrs[1:] :
    dic = {}
    tds = tr.find_all("td")
    
    if ']' in str(tds[1].text) :
        tds[1] = tds[1].text[:-3]
        tds[1] = tds[1].replace('[', '')
    else :
        tds[1] = tds[1].text
        
        
    dic['year'] = tds[0].text
    dic['game'] = tds[1]
    dic['genre'] = tds[2].text
    dic['flatform'] = tds[3].text
    dic['developer'] = tds[4].text
    gotyIgn.append(dic)
 
# DEST 차트

bsoup = BeautifulSoup(htmlText, "html.parser")
btab = bsoup.findAll("table", {"class" : "wikitable"})[4]    
btrs = btab.find("tbody").find_all("tr")
btdcols = btrs[1].find_all("td")

gotyDest = []

for tr in btrs[1:] :
    dic = {}
    tds = tr.find_all("td")
    
    if ']' in str(tds[1].text) :
        tds[1] = tds[1].text[:-3]
        tds[1] = tds[1].replace('[', '')
    else :
        tds[1] = tds[1].text
        
    dic['year'] = tds[0].text
    dic['game'] = tds[1]
    dic['genre'] = tds[2].text
    dic['flatform'] = tds[3].text
    dic['developer'] = tds[4].text
    gotyDest.append(dic)
    
# Edge 차트

bsoup = BeautifulSoup(htmlText, "html.parser")
btab = bsoup.findAll("table", {"class" : "wikitable"})[6]    
btrs = btab.find("tbody").find_all("tr")
btdcols = btrs[1].find_all("td")

gotyEdge = []

for tr in btrs[1:] :
    dic = {}
    tds = tr.find_all("td")
    
    if ']' in str(tds[1].text) :
        tds[1] = tds[1].text[:-3]
        tds[1] = tds[1].replace('[', '')
    else :
        tds[1] = tds[1].text
        
    dic['year'] = tds[0].text
    dic['game'] = tds[1]
    dic['genre'] = tds[2].text
    dic['flatform'] = tds[3].text
    dic['developer'] = tds[4].text
    gotyEdge.append(dic)
    
    
# Giant Bomb 차트

bsoup = BeautifulSoup(htmlText, "html.parser")
btab = bsoup.findAll("table", {"class" : "wikitable"})[8]    
btrs = btab.find("tbody").find_all("tr")
btdcols = btrs[1].find_all("td")

goty_Giant_Bomb = []

for tr in btrs[1:] :
    dic = {}
    tds = tr.find_all("td")
    
    if ']' in str(tds[1].text) :
        tds[1] = tds[1].text[:-4]
        tds[1] = tds[1].replace('[', '')
    else :
        tds[1] = tds[1].text
        
    dic['year'] = tds[0].text
    dic['game'] = tds[1]
    dic['genre'] = tds[2].text
    dic['flatform'] = tds[3].text
    dic['developer'] = tds[4].text
    goty_Giant_Bomb.append(dic)
 

# GUI 구현

window = Tk()

style = ttk.Style()

style.theme_use("clam")

style.configure("Treeview",
               background="silver",
               foreground="black",
               fieldbackground="silver")


style.map('Treeview', background=[('selected', 'azure4')])


window.geometry("850x650")
window.resizable(0,0)
window['bg'] = 'sky blue'


treeTable = ttk.Treeview(window)
treeTable["columns"] = ("year", "game", "genre", "flatform", "developer")
treeTable.column("#0", width = 20, anchor = "center")
treeTable.column("year", width = 50, anchor = "center")
treeTable.column("game", width = 200, anchor = "center")
treeTable.column("genre", width = 150, anchor = "center")
treeTable.column("flatform", width = 100, anchor = "center")
treeTable.column("developer", width = 120, anchor = "center")


treeTable.heading("#0", text="No.", anchor = "center") 
treeTable.heading("year", text="년도", anchor = "center")
treeTable.heading("game", text="게임명", anchor = "center")
treeTable.heading("genre", text="장르", anchor = "center")
treeTable.heading("flatform", text="플랫폼", anchor = "center")
treeTable.heading("developer", text="개발사", anchor = "center")

        
IgnButton = Button(window, text="IGN 차트", 
            relief='ridge',
            font=("Helvetica 12 bold"),
            cursor="target",
            command=call_Ign)

DestButton = Button(window, text="디스트럭토이드 차트", 
            relief='ridge',
            font=("Helvetica 12 bold"),
            cursor="target",
            command=callDest)

EdgeButton = Button(window, text="에지 차트", 
            relief='ridge',
            font=("Helvetica 12 bold"),
            cursor="target",
            command=callEdge)

GiantButton = Button(window, text="자이언트 밤 차트", 
            relief='ridge',
            font=("Helvetica 12 bold"),
            cursor="target",
            command=callGiant)
    
GiantButton.pack(side="bottom", fill="both")
EdgeButton.pack(side="bottom", fill="both") 
DestButton.pack(side="bottom", fill="both")  
IgnButton.pack(side="bottom", fill="both")
  
    
setTableItemIGN()

treeTable.place(x = 10, y = 90, width = 830, height = 400)
window.mainloop()
```
